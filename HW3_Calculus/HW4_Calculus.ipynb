```python
def get_min_with_gd(x0, N, lr, threshold):    #Решение с использованием напрямую производной f(x) (так как у нас одномерный случай)
    x = x0
    history = [x]
    for i in range(N):
      k = df(x)
      x1 = x - lr * k
      history.append(x1)
      if np.abs(x1 - x) < threshold:
          break
      x = x1
    return np.array(history)
```

```python
x0 = 5.0
N = 1000
lr = 0.01
threshold = 0.0000000001
history = get_min_with_gd(x0, N, lr, threshold)

history
```

```python
def get_min_with_gd(x0, func_grad, N, lr, threshold):   # Решение с использованием функции func_grad
    x = np.array([x0])
    history = [x]
    for i in range(N):
      grad = func_grad(x)
      norm_grad = np.linalg.norm(grad)
      x1 = x -lr* grad / norm_grad
      history.append(x1)
      if np.abs(x1 - x) < threshold:
          break
      x = x1
    return np.array(history)
```

**Аналитическое вычисление экстремума функции h(x, y) и определение его типа (максимум или минимум) по критерию Сильвестра**

Вычисление частных производных первого порядка: $df/dx=(ln((e^x + e^{-2x}) \cdot(e^{y + 1} + 0.5\cdot e ^{-y})))'=  (e^x + e^{-2x})' \cdot(e^{y + 1} + 0.5\cdot e ^{-y})+(e^x + e^{-2x}) \cdot(e^{y + 1} + 0.5\cdot e ^{-y})'/(e^x + e^{-2x}) \cdot(e^{y + 1} + 0.5\cdot e ^{-y})=(e^x -2 \cdot e^{-2x})/(e^x + e^{-2x})$
$df/dy=(ln((e^x + e^{-2x}) \cdot(e^{y + 1} + 0.5\cdot e ^{-y})))'=  (e^x + e^{-2x})' \cdot(e^{y + 1} + 0.5\cdot e ^{-y})+(e^x + e^{-2x}) \cdot(e^{y + 1} + 0.5\cdot e ^{-y})'/(e^x + e^{-2x}) \cdot(e^{y + 1} + 0.5\cdot e ^{-y})= (e^{y + 1} - 0.5\cdot e ^{-y})/(e^{y + 1} + 0.5\cdot e ^{-y})$.  
Приравнивая обе частные производные к нулю, найдем координаты точки возможного экстремума. Так как обе частные производные представляют собой дроби, то нулю они могут быть равны. только если их числитель равне нулю:
$e^x -2 \cdot e^{-2x} = 0$, $(e^{3x} -2)/e^{2x}=0$;  $e^{3x} = 2$; $3x=ln2$; $x = ln2/3$; $x\approx0,2310490602$.

$(e^{y + 1} - 0.5\cdot e ^{-y})=0$; $(e^{2y + 1} - 0.5)/ e ^{y}=0$; $e^{2y + 1} = 0.5$; $2y+1=ln0.5$; $y=(ln0.5-1)/2$; $y\approx-0.84657359$ .
Значит, в точке с приблизительными координатами $(0,2310490602, -0.84657359)$ функция $f(x, y)$ имеет единственный вероятный экстремум.

Далее вычисляются частные производные второго порядка $d^2g/d^2x$;    $d^2g/d^2y$;    $d^2g/dxdy$;
   $d^2g/dydx$ и находятся их значения в точке возможного экстремума.

Так как матрица Гессе в данной точке положительно определена (частные вторые частные производные по x и y больше нуля, а смешанные частные производные равны нулю, значит первый и второй угловые миноры будут положительны), значит в найденной точке экстремума находится минимум функции.
Так как у функции в найденной точке - единственный экстремум, то он является глобальным экстремумом.

**Аналитическое вычисление экстремума функции g(x, y) и определение его типа (максимум или минимум) по критерию Сильвестра**

Вычисление частных производных первого порядка: $dg/dx=(-x^2-y^2)'=(-x^2)'=-2x$,$dg/dy=(-x^2-y^2)'=(-y^2)'=-2y$.  
Приравнивая обе частные производные к нулю, найдем координаты точки возможного экстремума (0, 0).
Все частные производные второго порядка - это константы:
$d^2g/d^2x=(-2x)'=-2$;    $d^2g/d^2y=(-2y)'=-2$;    $d^2g/dxdy=(-2x)'=0$;
   $d^2g/dydx=(-2y)'=0$.

Так как угловой минор первого порядка меньше нуля, а второго порядка - больше нуля, то матрица Гессе - отрицательно определена (угловой минор первого порядка равен *$d^2g/d^2x=(-2x)'=-2$*, угловой минор второго порядка положителен, т.к.равен определителю второго порядка из вторых производных и равен 4 $((-2)\cdot(-2)-0\cdot0)=4)$, а значит в найденной точке экстремума находится максимум функции.
Так как у функции в найденной точке - единственный экстремум, то он является глобальным максимумом.

Реализуйте функцию `get_min_with_rmsprop`, которая принимает на вход начальную точку $x_0$, число шагов, `lr`, `beta` и `threshold`; внутри себя производит поиск минимума при помощи градиентного спуска с модификацией RMSProp; и возвращает последовательность точек, полученных на каждом шаге.

Напомним, как выглядят формулы для этого алгоритма:

$$g^{n+1} = ∇f(w^{n})$$
$$\gamma^{n+1} = (1 - \beta) \gamma^{n} + \beta (g^{n+1}) ^ 2$$
$$w^{n+1} = w^n - \frac{\alpha}{\sqrt{\gamma^{n+1} + ɛ}}⋅g^{n+1}$$

Здесь $α$ — learning rate, $ɛ$ — малое число (можно взять 0.0001 — оно нужно, чтобы случайно не было деления на 0), $\beta$ — мера учета инерции.



```python
def get_min_with_rmsprop(x0, func_grad, N, lr, beta, threshold):
    x = np.array(x0)
    history = [x.copy()]
    avg_sq_grad = np.zeros_like(x)

    for i in range(N):
        grad = func_grad(x)
        avg_sq_grad = beta * avg_sq_grad + (1 - beta) * grad**2+beta * avg_sq_grad
        x1 = x - lr * grad / (np.sqrt(avg_sq_grad) + 0.0001)
        history.append(x1.copy())

        if np.linalg.norm(x1 - x) < threshold:
            break

        x = x1

    return np.array(history)
```

**Рассмотрим применение метода градиентного спуска для функции 5 переменных $f(x, y, z, p, q)=x^2+y^2+z^2+p^2+q^2$.** Данная функция имеет единственный экстремум - глобальный минимум в точке (0,0,0,0,0). Убедимся в этом.

```python
def fffff(xyzpq):
  x, y, z, p, q = xyyzpq

  return np.array(x**2+y**2+z**2+p**2+q**2)
```

```python
def fffff(xyzpq):                       #Реализация подсчета значения функции f(x, y, z, p, q) в точке
  x = xyzpq[0]
  y = xyzpq[1]
  z = xyzpq[2]
  p = xyzpq[3]
  q = xyzpq[4]
  return np.array([x**2+y**2+z**2+p**2+q**2])
```

```python
def grfffff(xyzpq):                       #Реализация подсчета значения градиента f(x, y, z, p, q) в точке
  x = xyzpq[0]
  y = xyzpq[1]
  z = xyzpq[2]
  p = xyzpq[3]
  q = xyzpq[4]
  return np.array([2*x, 2*y, 2*z, 2*p, 2*q])
```

```python
print(f"В точке x = {1}, y = {1}, z = {1}, p = {1}, q = {1}, градиент функции f(x, y, z, p, q) принимает значение {grfffff([1, 1, 1, 1, 1])}")
```

```python
def get_min_with_rmsprop1(x0, func_grad, N, lr, beta, threshold):
    x = np.array(x0)
    history = [x.copy()]
    avg_sq_grad = np.zeros_like(x)

    for i in range(N):
        grad = func_grad(x)
        avg_sq_grad = beta * avg_sq_grad + (1 - beta) * grad**2+beta * avg_sq_grad
        x1 = x - lr * grad / (np.sqrt(avg_sq_grad) + 0.0001)
        history.append(x1.copy())

        if np.linalg.norm(x1 - x) < threshold:
            break

        x = x1

    return np.array(history)
```
